```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# === Slicing input ===
slicing_directions = ['H', 'V', 'V', 'H', 'H', 'V', 'V', 'H']  # for 9 facilities
facility_dimensions = {
    'A': (8, 16), 'B': (7, 30), 'C': (10, 30), 'D': (7, 46), 'E': (6, 46),
    'F': (16, 9), 'G': (6, 14), 'H': (17, 14), 'I': (7, 14)
}
layout_width = 100
layout_height = 100

# === Distance function ===
def euclidean(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))

# === Objective function ===
def compute_layout_cost(facility_rects, flow_matrix, cost_matrix):
    layout = [f[0] for f in facility_rects]
    centers = {name: (x + w / 2, y + h / 2) for name, (x, y, w, h) in facility_rects}
    total_cost = 0
    for i in range(len(layout)):
        for j in range(len(layout)):
            if i != j:
                d = euclidean(centers[layout[i]], centers[layout[j]])
                total_cost += flow_matrix[i][j] * d * cost_matrix[i][j]
    return total_cost

# === Slicing logic ===
def recursive_slicing_weighted(x, y, width, height, facilities, cuts, cut_index=0):
    if len(facilities) == 1:
        return [(facilities[0], (x, y, width, height))]
    cut = cuts[cut_index]
    mid = len(facilities) // 2
    left, right = facilities[:mid], facilities[mid:]
    area = lambda f: facility_dimensions[f][0] * facility_dimensions[f][1]
    la, ra = sum(area(f) for f in left), sum(area(f) for f in right)
    total = la + ra
    result = []
    if cut == 'H':
        h1 = height * (la / total)
        result += recursive_slicing_weighted(x, y + (height - h1), width, h1, left, cuts, cut_index + 1)
        result += recursive_slicing_weighted(x, y, width, height - h1, right, cuts, cut_index + 1 + len(left) - 1)
    else:
        w1 = width * (la / total)
        result += recursive_slicing_weighted(x, y, w1, height, left, cuts, cut_index + 1)
        result += recursive_slicing_weighted(x + w1, y, width - w1, height, right, cuts, cut_index + 1 + len(left) - 1)
    return result

# === Flow and Cost Matrices (sample for 9 facilities) ===
flow_matrix = np.array([
    [0, 10, 5, 7, 2, 3, 0, 1, 4],
    [10, 0, 8, 1, 4, 0, 2, 1, 3],
    [5, 8, 0, 2, 6, 7, 1, 0, 2],
    [7, 1, 2, 0, 9, 3, 0, 2, 1],
    [2, 4, 6, 9, 0, 5, 1, 3, 2],
    [3, 0, 7, 3, 5, 0, 6, 2, 1],
    [0, 2, 1, 0, 1, 6, 0, 4, 3],
    [1, 1, 0, 2, 3, 2, 4, 0, 5],
    [4, 3, 2, 1, 2, 1, 3, 5, 0]
])
cost_matrix = np.ones((9, 9))  # Uniform cost

# === Function to evaluate multiple layouts ===
def evaluate_layouts(layouts, slicing_directions, flow_matrix, cost_matrix):
    results = []
    for layout in layouts:
        rects = recursive_slicing_weighted(0, 0, layout_width, layout_height, layout, slicing_directions)
        cost = compute_layout_cost(rects, flow_matrix, cost_matrix)
        results.append((layout, cost))
    return sorted(results, key=lambda x: x[1])

# === Example: User-provided layouts ===
user_layouts = [
    ['D', 'A', 'H', 'B', 'E', 'F', 'I', 'C', 'G'],
    ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    ['G', 'F', 'E', 'D', 'C', 'B', 'A', 'I', 'H'],
    # Add more layouts as needed
]

# Evaluate user-provided layouts
evaluated_results = evaluate_layouts(user_layouts, slicing_directions, flow_matrix, cost_matrix)

# Display results
for i, (layout, cost) in enumerate(evaluated_results, 1):
    print(f"{i:02}: Layout {layout} → Cost: {cost:.2f}")
```

MUTATION

```python
import random

# Assume `offspring_layouts` contains the 10 offspring from the crossover step
# Example:
# offspring_layouts = [
#     ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
#     ...
# ]

def swap_mutation(chromosome, mutation_rate=0.2):
    mutated = chromosome.copy()
    if random.random() < mutation_rate:
        i, j = random.sample(range(len(mutated)), 2)
        mutated[i], mutated[j] = mutated[j], mutated[i]
    return mutated

# Apply mutation to all offspring
mutated_offspring = [swap_mutation(child) for child in offspring_layouts]

# Display mutated results
for idx, layout in enumerate(mutated_offspring, 1):
    print(f"Mutated Offspring {idx}: {layout}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# === Slicing input ===
slicing_directions = ['H', 'V', 'V', 'H', 'H', 'V', 'V', 'H']  # for 9 facilities
facility_dimensions = {
    'A': (8, 16), 'B': (7, 30), 'C': (10, 30), 'D': (7, 46), 'E': (6, 46),
    'F': (16, 9), 'G': (6, 14), 'H': (17, 14), 'I': (7, 14)
}
layout_width = 100
layout_height = 100

# === Distance function ===
def euclidean(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))

# === Objective function ===
def compute_layout_cost(facility_rects, flow_matrix, cost_matrix):
    layout = [f[0] for f in facility_rects]
    centers = {name: (x + w / 2, y + h / 2) for name, (x, y, w, h) in facility_rects}
    total_cost = 0
    for i in range(len(layout)):
        for j in range(len(layout)):
            if i != j:
                d = euclidean(centers[layout[i]], centers[layout[j]])
                total_cost += flow_matrix[i][j] * d * cost_matrix[i][j]
    return total_cost

# === Slicing logic ===
def recursive_slicing_weighted(x, y, width, height, facilities, cuts, cut_index=0):
    if len(facilities) == 1:
        return [(facilities[0], (x, y, width, height))]
    cut = cuts[cut_index]
    mid = len(facilities) // 2
    left, right = facilities[:mid], facilities[mid:]
    area = lambda f: facility_dimensions[f][0] * facility_dimensions[f][1]
    la, ra = sum(area(f) for f in left), sum(area(f) for f in right)
    total = la + ra
    result = []
    if cut == 'H':
        h1 = height * (la / total)
        result += recursive_slicing_weighted(x, y + (height - h1), width, h1, left, cuts, cut_index + 1)
        result += recursive_slicing_weighted(x, y, width, height - h1, right, cuts, cut_index + 1 + len(left) - 1)
    else:
        w1 = width * (la / total)
        result += recursive_slicing_weighted(x, y, w1, height, left, cuts, cut_index + 1)
        result += recursive_slicing_weighted(x + w1, y, width - w1, height, right, cuts, cut_index + 1 + len(left) - 1)
    return result

# === Flow and Cost Matrices (sample for 9 facilities) ===
flow_matrix = np.array([
    [0, 10, 5, 7, 2, 3, 0, 1, 4],
    [10, 0, 8, 1, 4, 0, 2, 1, 3],
    [5, 8, 0, 2, 6, 7, 1, 0, 2],
    [7, 1, 2, 0, 9, 3, 0, 2, 1],
    [2, 4, 6, 9, 0, 5, 1, 3, 2],
    [3, 0, 7, 3, 5, 0, 6, 2, 1],
    [0, 2, 1, 0, 1, 6, 0, 4, 3],
    [1, 1, 0, 2, 3, 2, 4, 0, 5],
    [4, 3, 2, 1, 2, 1, 3, 5, 0]
])
cost_matrix = np.ones((9, 9))  # Uniform cost

# === Example: Evaluate a batch of layouts ===
population = [
    ['D','A','H','B','E','F','I','C','G'],
    ['A','B','C','D','E','F','G','H','I'],
    ['G','F','E','D','C','B','A','I','H'],
    # ... more layouts
]

# Evaluate cost for each layout using slicing
results = []
for layout in population:
    rects = recursive_slicing_weighted(0, 0, layout_width, layout_height, layout, slicing_directions)
    cost = compute_layout_cost(rects, flow_matrix, cost_matrix)
    results.append((layout, cost))

# Display results sorted by cost
for i, (layout, cost) in enumerate(sorted(results, key=lambda x: x[1]), 1):
    print(f"{i:02}: Layout {layout} → Cost: {cost:.2f}")
```
